```python
import pandas as pd
import psycopg2

# Set up your PostgreSQL connection
connection = psycopg2.connect(
    host="localhost",
    database="telecom_data",  # Replace with your database name
    user="postgres",  # Replace with your username
    password="1040"  # Replace with your password
)

# Define the SQL query to get the top 10 rows based on a column (e.g., total data usage)
query = """
SELECT 
    "IMSI", 
    AVG("Dur. (ms)") AS avg_duration,
    SUM("Total DL (Bytes)" + "Gaming UL (Bytes)" + "Other DL (Bytes)" + "Other UL (Bytes)" + "Total UL (Bytes)") AS total_data_usage,
    AVG("Avg RTT DL (ms)") AS avg_rtt_dl,
    AVG("Avg RTT UL (ms)") AS avg_rtt_ul
FROM 
    xdr_data
GROUP BY
    "IMSI"
ORDER BY
    total_data_usage DESC
LIMIT 10;
"""

# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)

# Display the result
print(df)

# Close the connection
connection.close()
```

```python
query = """
WITH ranked_handsets AS (
    SELECT 
        "Handset Manufacturer", 
        "IMEI",
        COUNT("IMEI") AS handset_count,
        ROW_NUMBER() OVER (PARTITION BY "Handset Manufacturer" ORDER BY COUNT("IMEI") DESC) AS rank
    FROM 
        xdr_data
    GROUP BY 
        "Handset Manufacturer", "IMEI"
)
SELECT 
    "Handset Manufacturer", 
    "IMEI", 
    handset_count
FROM 
    ranked_handsets
WHERE 
    rank <= 5
ORDER BY 
    "Handset Manufacturer", rank;
"""

# Execute the query and load the result into a DataFrame
df_top_5_handsets_per_manufacturer = pd.read_sql_query(query, connection)

# Show the result
print(df_top_5_handsets_per_manufacturer)
```
